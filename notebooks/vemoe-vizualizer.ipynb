```python
moe_lens_address = "0x527eF5B98337b532F0215bDb0DA709439Ad3B0F6"
rewarder_address = "0x628Be0a99096e8af0AAD49746B1818Eb6da588e7"
farm_pid = 29
rewarder_pid = 735
wallet_address = "0x008eEB180F9Fb7190b8f2C0Ed1bEEf92Fb2Ddf32"

start_block = 65121644
# delta = 100
delta = 5000

moe_lens = w3.eth.contract(address=moe_lens_address, abi=moe_lens_abi)
rewarder = w3.eth.contract(address=rewarder_address, abi=rewarder_abi)
```

```python
def getVoteDataAt(block):
    veMoeData = moe_lens.functions.getVeMoeData(farm_pid, 1, wallet_address).call(block_identifier=block)
    data = veMoeData[8][0]

    return {
        "date": w3.eth.get_block(block)["timestamp"],
        "block": block,
        "total_votes": veMoeData[2] / 1e18,
        "total_weights": veMoeData[3] / 1e18,
        "votes_on_farm": data[1] / 1e18,
        "weights_on_farm": data[2] / 1e18,
        "staked_on_rewarder": data[3][5] / 1e18,
        "staked_wallet": data[4] / 1e18,
    }
```

```python
latest_block = w3.eth.block_number

# read the csv
try:
    df = pd.read_csv("vemoe_data.csv")
    start_block = df["block"].max() + delta
except:
    df = pd.DataFrame()

if start_block <= latest_block:
    print(f"Starting from block {start_block} to {latest_block}, {int(((latest_block - start_block) + 1 )/ delta)} iterations")

    # get the data
    data = []
    for block in range(start_block, latest_block, delta):
        data.append(getVoteDataAt(block))

    # convert the data to a dataframe
    if data:
        new_df = pd.DataFrame(data)
        new_df["date"] = pd.to_datetime(new_df["date"], unit="s").dt.tz_localize('UTC').dt.tz_convert('Europe/Paris')

        # append the new data to the csv, fixing the index
        df = pd.concat([df, new_df], ignore_index=True)

        df.to_csv("vemoe_data.csv", index=False)
        df = pd.read_csv("vemoe_data.csv")
        df["date"] = pd.to_datetime(df["date"])
```

```python
date = df["date"]
vote_allocations = df["votes_on_farm"] * 100 / df["total_votes"]
weight_allocations = df["weights_on_farm"] * 100 / df["total_weights"]
wallet_vote_allocations = df["staked_wallet"] * 100 / df["total_votes"]

fig, ax = plt.subplots()
plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%d/%m %H:%m"))

ax.set_title("Allocations on farm (in %)")
ax.plot(date, vote_allocations, label="Votes")
ax.plot(date, weight_allocations, label="Weights")
ax.plot(date, wallet_vote_allocations, label="Wallet votes")
ax.legend()
ax.grid(True)

ax.xaxis.set_major_locator(plt.MaxNLocator(4))

plt.show()
```

```python
date = df["date"]
staked_allocations = df["staked_on_rewarder"] * 100 / df["votes_on_farm"]
staked_wallet = df["staked_wallet"] * 100 / df["staked_on_rewarder"]
staked_wallet = staked_wallet.fillna(0)

fig, ax = plt.subplots()
plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%d/%m %H:%m"))

ax.set_title("Staked on rewarder (in %)")
ax.plot(date, staked_allocations, label="Votes staked")
ax.plot(date, staked_wallet, label="Staked by wallet")
ax.legend()
ax.grid(True)

ax.xaxis.set_major_locator(plt.MaxNLocator(4))

plt.show()
```
